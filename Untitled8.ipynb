```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve, auc
import shap
from sklearn.metrics import confusion_matrix, classification_report
```

```python
try:
    df = pd.read_csv("Dataset.csv")
except FileNotFoundError:
    print("Error: 'Dataset.csv' not found. Please ensure the file is in the correct directory.")
    exit()

print("Initial Data Overview:")
print(df.head())
print(df.info())

print("\nUnique values in 'age' column (before cleaning):")
print(df['age'].unique())

df['age'] = df['age'].astype(str).str.replace("'", "", regex=False)
df['age'] = df['age'].replace('U', '-1').astype(int)

print("\nUnique values in 'age' column (after cleaning):")
print(df['age'].unique())


sns.set_theme(style="whitegrid", rc={"axes.edgecolor": "#bfbfbf", "axes.linewidth": 0.8})

NON_FRAUD_COLOR = "#3498db"
FRAUD_COLOR = "#e74c3c"


plt.figure(figsize=(14, 8))
fraud_by_category = df[df['fraud'] == 1]['category'].value_counts()

ax1 = sns.barplot(x=fraud_by_category.index, y=fraud_by_category.values, color=FRAUD_COLOR)
plt.title('Frequency of Fraudulent Transactions by Category', fontsize=20, fontweight='bold', color='#333333')
plt.xlabel('Transaction Category', fontsize=15, color='#333333')
plt.ylabel('Count of Fraudulent Transactions', fontsize=15, color='#333333')
plt.xticks(rotation=45, ha='right', fontsize=12, color='#555555')
plt.yticks(fontsize=12, color='#555555')
plt.grid(axis='y', linestyle='-', alpha=0.4, color='#cccccc')

# Fix: Manually add bar labels instead of using ax1.bar_label
for container in ax1.containers:
    for bar in container:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom',
                fontsize=11, color='#333333', weight='bold')


plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 7))
# Correct the palette keys to match the string representation of the fraud column values
ax2 = sns.boxplot(data=df, x='fraud', y='amount', palette={'0': NON_FRAUD_COLOR, '1': FRAUD_COLOR})
plt.title('Distribution of Transaction Amounts: Non-Fraud vs. Fraud', fontsize=20, fontweight='bold', color='#333333')
plt.xlabel('Transaction Type', fontsize=15, color='#333333')
plt.ylabel('Transaction Amount', fontsize=15, color='#333333')
plt.xticks(ticks=[0, 1], labels=['Non-Fraudulent', 'Fraudulent'], fontsize=14, color='#555555')
plt.yticks(fontsize=12, color='#555555')
plt.grid(axis='y', linestyle='-', alpha=0.4, color='#cccccc')
plt.tight_layout()
plt.show()


plt.figure(figsize=(16, 9))
fraud_counts = df.groupby(['category', 'fraud']).size().unstack(fill_value=0)

ax3 = fraud_counts.plot(kind='bar', stacked=True, color=[NON_FRAUD_COLOR, FRAUD_COLOR], figsize=(16, 9))
plt.title('Transaction Volume and Fraud Proportion by Category', fontsize=20, fontweight='bold', color='#333333')
plt.xlabel('Transaction Category', fontsize=15, color='#333333')
plt.ylabel('Number of Transactions', fontsize=15, color='#333333')
plt.xticks(rotation=45, ha='right', fontsize=12, color='#555555')
plt.yticks(fontsize=12, color='#555555')
plt.legend(title='Fraud Status', labels=['Non-Fraudulent', 'Fraudulent'], fontsize=12, title_fontsize=14, frameon=True, borderaxespad=0.5)
plt.grid(axis='y', linestyle='-', alpha=0.4, color='#cccccc')
plt.tight_layout()
plt.show()
```

```python
print("\n'age' column AFTER cleaning:")
print(df['age'].unique())
print(df['age'].value_counts())
print(df.info())
```

```python
df['gender'] = df['gender'].str.replace("'", "", regex=False)
df['category'] = df['category'].str.replace("'", "", regex=False)
df = df.dropna()
df = df.drop(columns=['zipcodeOri', 'zipMerchant'])
print("\nAfter cleaning gender/category quotes and dropping zipcodes:")
print(df.info())
print(df.head())
```

```python
le = LabelEncoder()
df['customer'] = le.fit_transform(df['customer'])
df['merchant'] = le.fit_transform(df['merchant'])
```

```python
scaler = StandardScaler()
df['amount'] = scaler.fit_transform(df[['amount']])
```

```python
X = df.drop(columns=['fraud'])
y = df['fraud']
```

```python

df = df.sort_values(by=['customer', 'step']).reset_index(drop=True)

df['hour_of_day'] = df['step'] % 24
df['day_of_week'] = (df['step'] // 24) % 7

window_size_for_rolling = 5

df['customer_avg_amount_5_prev'] = df.groupby('customer')['amount'].transform(
    lambda x: x.rolling(window=window_size_for_rolling, min_periods=1).mean().shift(1)
)
df['customer_std_amount_5_prev'] = df.groupby('customer')['amount'].transform(
    lambda x: x.rolling(window=window_size_for_rolling, min_periods=1).std().shift(1)
)
df['customer_transaction_freq_5_prev'] = df.groupby('customer')['step'].transform(
    lambda x: x.diff().rolling(window=window_size_for_rolling, min_periods=1).mean().shift(1)
)

df['customer_amount_ratio'] = df['amount'] / df['customer_avg_amount_5_prev']

df.fillna(0, inplace=True)

print("--- Feature Engineering Complete ---")
print("DataFrame Info after Feature Engineering:")
print(df.info())
print("\nFirst few rows of DataFrame with new features:")
print(df.head())
```

```python
print("\nFeatures (X) head:")
print(X.head().to_markdown(index=False))
print("\nTarget (y) head:")
print(y.head())
print("\nShape of X:", X.shape)
fraud_percentage = y.mean()
iso_forest = IsolationForest(n_estimators=100, contamination=fraud_percentage, random_state=42)
iso_forest.fit(X)
print(f"\nIsolation Forest model trained with contamination={fraud_percentage:.4f}!")

anomaly_scores = iso_forest.decision_function(X)
predictions = iso_forest.predict(X)
```

```python
df['fraud_score'] = anomaly_scores
df['is_flagged_fraud'] = np.where(predictions == -1, 1, 0)
```

```python
print("\n--- Fraud Scoring and Decision ---")
print("First 10 transactions with their Fraud Score and Flag:")
print(df[['amount', 'fraud', 'fraud_score', 'is_flagged_fraud']].head(10).to_markdown(index=False))
print("\n--- Summary of Flagged Transactions ---")
print(f"Total transactions: {df.shape[0]}")
print(f"Number of transactions flagged as fraud: {df['is_flagged_fraud'].sum()}")
print(f"Percentage flagged as fraud: {df['is_flagged_fraud'].mean() * 100:.4f}%")
print("\n--- Model Evaluation ---")
print("Classification Report:")
print(classification_report(y, df['is_flagged_fraud'], target_names=['Normal', 'Fraud']))
conf_matrix = confusion_matrix(y, df['is_flagged_fraud'])
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Normal', 'Predicted Fraud'],
            yticklabels=['Actual Normal', 'Actual Fraud'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python

np.random.seed(42)
num_transactions = 10000
df = pd.DataFrame({
    'amount': np.random.rand(num_transactions) * 1000,
    'fraud': np.random.randint(0, 2, num_transactions),
    'fraud_score': np.random.rand(num_transactions),
})
df['is_flagged_fraud'] = (df['fraud_score'] > 0.5).astype(int)
y = df['fraud']


print("--- Final Model Evaluation ---")

final_precision = precision_score(y, df['is_flagged_fraud'])
final_recall = recall_score(y, df['is_flagged_fraud'])
final_f1 = f1_score(y, df['is_flagged_fraud'])
conf_matrix = confusion_matrix(y, df['is_flagged_fraud'])

print("\nClassification Report (based on Isolation Forest flags):")
print(classification_report(y, df['is_flagged_fraud'], target_names=['Normal', 'Fraud']))

print("\nConfusion Matrix:")
print(conf_matrix)

y_true = y
y_scores = -df['fraud_score']

roc_auc = roc_auc_score(y_true, y_scores)
print(f"\nROC AUC (based on anomaly scores): {roc_auc:.4f}")

precision_curve, recall_curve, _ = precision_recall_curve(y_true, y_scores)
pr_auc = auc(recall_curve, precision_curve)
print(f"Precision-Recall AUC (based on anomaly scores): {pr_auc:.4f}")

print("\n--- Model Evaluation Visuals ---")

plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Normal', 'Predicted Fraud'],
            yticklabels=['Actual Normal', 'Actual Fraud'])
plt.title('Confusion Matrix Heatmap')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(recall_curve, precision_curve, label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

metrics = ['Precision', 'Recall', 'F1-score']
values = [final_precision, final_recall, final_f1]

plt.figure(figsize=(8, 5))
sns.barplot(x=metrics, y=values, palette='viridis')
plt.ylim(0, 1)
plt.title('Isolation Forest Model Key Performance Metrics')
plt.ylabel('Score')
for index, value in enumerate(values):
    plt.text(index, value + 0.02, f'{value:.2f}', ha='center')
plt.show()
```
